# Execution Volume Impact Modeling
Simulates transaction cost curves using Kyle's Lambda and the square-root law to model price impact based on order size.


```python
import numpy as np
import matplotlib.pyplot as plt

order_sizes = np.linspace(100, 50000, 100)
daily_volume = 1000000
volatility = 0.02

# Square-root law of market impact
impact = volatility * np.sqrt(order_sizes / daily_volume) * 10000 # in bps

plt.figure(figsize=(10, 4))
plt.plot(order_sizes, impact, color="orange", lw=2)
plt.title("Slippage Price Impact Model (Bps)")
plt.xlabel("Order Size (Units)")
plt.ylabel("Estimated Slippage (Bps)")
plt.grid(True, alpha=0.3)
plt.show()

```
