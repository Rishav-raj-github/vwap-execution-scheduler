# VWAP Execution Scheduler Profile
This notebook visualizes the standard U-shaped intraday volume profile and demonstrates how a large execution block (e.g., 100,000 units) is scheduled across the intervals to minimize venue footprint.


```python
import numpy as np
import matplotlib.pyplot as plt

# 8 intervals representing the trading day
intervals = np.arange(1, 9)
historical_fractions = np.array([0.20, 0.12, 0.08, 0.05, 0.05, 0.10, 0.15, 0.25])

# Total order size
total_order_size = 100000
scheduled_slices = total_order_size * (historical_fractions / np.sum(historical_fractions))

print("Intraday Execution Schedule:")
for i, slice_size in enumerate(scheduled_slices):
    print(f" Interval {i+1} | Target Execution: {int(slice_size):6d} Units")

```
