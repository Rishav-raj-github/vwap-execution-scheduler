# VWAP vs TWAP Slippage Comparison
Compares the average execution slippage between VWAP and TWAP under simulated daily volume distribution shapes.


```python
import numpy as np

# U-shaped volume curve
vol_profile = np.array([0.25, 0.10, 0.08, 0.07, 0.08, 0.12, 0.30])
prices = np.array([100.10, 100.20, 100.15, 100.05, 100.10, 100.25, 100.40])

# VWAP Execution slices
vwap_qty = 10000 * vol_profile
vwap_cost = np.sum(vwap_qty * prices) / np.sum(vwap_qty)

# TWAP Execution slices (flat distribution)
twap_qty = np.ones_like(vol_profile) * (10000 / len(vol_profile))
twap_cost = np.sum(twap_qty * prices) / np.sum(twap_qty)

print(f"VWAP Weighted Cost: ${vwap_cost:.4f}")
print(f"TWAP Weighted Cost: ${twap_cost:.4f}")
print(f"VWAP Improvement: {(twap_cost - vwap_cost)*100:.2f} cents/unit")

```
